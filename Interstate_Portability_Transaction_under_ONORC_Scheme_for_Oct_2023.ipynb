```python
#let's look at the sum of transactions
data['txn_count'].sum()
```

```python
#let's find out 'txn_count' for the different home states 
data.groupby('homestatename')['txn_count'].sum()
```

```python
#Lets find out 'txn_count' for the different Sale State
grouped_sale_data=data.groupby('salestatename')['txn_count'].sum()
grouped_sale_data
```

```python
#Top 10 states with respect to 'txn_count' by 'salestatename'
top_10_salestate=grouped_sale_data.sort_values(ascending=False).head(10)
```

```python
top_10_salestate
```

```python
plt.barh(top_10_salestate.index,top_10_salestate.values,color='aqua')
plt.xlabel('Sum of Txn_Count')
plt.ylabel('States')
plt.title('Top 10 Selling States')
plt.show()
```

```python
#Top 10 states with respect to 'txn_count' by 'homestatename'
grouped_home_data=data.groupby('homestatename')['txn_count'].sum()
```

```python
grouped_home_data
```

```python
top_10_homestate=grouped_home_data.sort_values(ascending=False).head(10)
```

```python
top_10_homestate
```

```python
plt.barh(top_10_homestate.index,top_10_homestate.values,color='salmon')
plt.xlabel('Sum of Txn_Count')
plt.ylabel('States')
plt.title('Top 10 Home States')
plt.show()
```
